# How to perform a MERGE Operation b/w two DataFrame ?

```python
import pandas as pd
player = ['Player1', 'Player2', 'Player3']
point = [8,9,6]
title = ['Game1', 'Game2', 'Game3']
df1 = pd.DataFrame({'Player':player, 'Points':point, 'Title':title})
df1
```

```python
player = ['Player1', 'Player5', 'Player6']
power = ['Punch', 'Kick', 'Elbow']
title = ['Game1', 'Game5', 'Game6']
df2 = pd.DataFrame({'Player':player, 'Power':power, 'Title':title})
df2
```

```python
#Inner Merge
df1.merge(df2, on='Player', how='inner')
```

```python
# By default we merege a 2 DataFrame it perform INNER Merge
df1.merge(df2)
```

```python
#Left Merge
df1.merge(df2, on='Player', how='left')
```

```python
#Right Merge
df1.merge(df2, on='Player', how='right')
```

```python
#Outer Merge
df1.merge(df2, on='Player', how='outer')
```

# How to perform JOIN operation in Pandas ?

```python
player = ['Player1', 'Player2', 'Player3']
point = [8,9,6]
title = ['Game1', 'Game2', 'Game3']
df3 = pd.DataFrame({'Player':player, 'Point':point, 'Title':title}, index = ['L1', 'L2', 'L3'])
df3
```

```python
player = ['Player1', 'Player5', 'Player6']
power = ['Punch', 'Kick', 'Elbow']
title = ['Game1', 'Game5', 'Game6']
df4 = pd.DataFrame({'Players':player, 'Powers':power, 'Titles':title},index = ['L2', 'L3', 'L4'])
df4
```

```python
# Inner Join
df3.join(df4, how='inner')
```

```python
# Left Join
df3.join(df4, how='left')
```

```python
# Right Join
df3.join(df4, how='right')
```

```python
# Outer Join
df3.join(df4, how='outer')
```

# How to concatinate 2 data in Pandas ?


```python
pd.concat([df3, df4])
```

# What is data Cleaning ?

```python
# Rename column
cars = cars.rename(columns={'Unnamed': 'MODEL'})
cars
```

```python
# Fill the null values with mean of the column
cars.qsec = cars.qsec.fillna(cars.qsec.mean())
cars
```

```python
# Drop Unwanted column
cars = cars.drop(columns=['S. No.'])
cars
# Previously we have 'S. No.' columns which is unused so we are drop that column.
```

```python
# Find co-relation matrix
df = cars[['mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']].corr()
df
#Shows varbles are corelated to other variables
```

```python
# Changing mpg from string to float
cars.mpg = cars.mpg.astype(float)
# See the cahnge
cars.info(null_counts = True)
```

```python
# Again Find co-relation including mpg
df = cars[['mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']].corr()
df
```
